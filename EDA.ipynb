```python
import pandas as pd

# ar_first_half = ar.loc[:round(ar.shape[0]/2)]
# ar_second_half = ar.loc[round(ar.shape[0]/2)+1:]
# ar_first_half.to_csv("AirlineReviews1.csv")
# ar_second_half.to_csv("AirlineReviews2.csv")

# Read seperated data sources
ar1 = pd.read_csv("AirlineReviews1.csv")
ar2 = pd.read_csv("AirlineReviews2.csv")
# Union datasets
ar = pd.concat([ar1, ar2], axis=0)
# Drop unnecassery column
ar = ar.drop(ar.columns[0], axis=1)
```

```python
# Top 10 occurences in database
ar.groupby('AirlineName').size().reset_index(name='count').sort_values(by='count', ascending=False).head(10)
```

```python
ar.
```

```python
# Group by 'AirlineName' and count occurrences, then sort in descending order
grouped_counts = ar.groupby('AirlineName').size().sort_values(ascending=False)

# Get top 10 airlines
top_10_airlines = grouped_counts.head(10).index

# Filter the DataFrame to only include rows for the top 10 airlines
top_10_df = ar[ar['AirlineName'].isin(top_10_airlines)]

# Select columns ending with 'Rating' or 'OverallScore' and calculate the average for each airline
average_ratings = top_10_df.filter(like='Rating').groupby(top_10_df['AirlineName']).mean()

average_ratings
```
